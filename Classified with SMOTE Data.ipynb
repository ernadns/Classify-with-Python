```python
#Klasifikasi
import numpy 
import numpy as np 
import matplotlib as mpl 
import matplotlib.pyplot as plt  
import optunity 
import optunity.metrics 
from scipy import stats
import sklearn
from sklearn.metrics import classification_report, confusion_matrix 
from __future__ import print_function 
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, KFold, ShuffleSplit, StratifiedShuffleSplit, RepeatedStratifiedKFold 
from sklearn.feature_selection import SelectPercentile, f_classif 
from sklearn.naive_bayes import BernoulliNB 
from sklearn.svm import LinearSVC, SVC 
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import GridSearchCV 
from sklearn.metrics import classification_report, accuracy_score 
from sklearn.utils import shuffle 
from sklearn import svm, datasets 
from sklearn.metrics import roc_curve, auc 
from scipy import interp 
from imblearn.over_sampling import SMOTE 
from sklearn import metrics 
import sklearn.svm 
```

```python
Y_new = pd.DataFrame.as_matrix(Y_nyasmote)
X_new, Y_new = pd.DataFrame.as_matrix(datasmoteX), Y_new 
```

```python
#Regresi Logistik Biner
kfold= StratifiedKFold(n_splits=4, shuffle=True)  
kfold.get_n_splits(X_new) 
kfold.get_n_splits(Y_new)
lr= LogisticRegression() 
akurasi=[]

for train, test in kfold.split(X_new, Y_new): 
            (X_new[train], X_new[test])
            (Y_new[train], Y_new[test])
            hitung = lr.fit(X_new[train], Y_new[train])
            predictrbl1 = lr.predict(X_new[train])
            predictrbl = lr.predict(X_new[test])
            fpr1, tpr1, _1 = metrics.roc_curve(Y_new[train], predictrbl1)
            fpr, tpr, _ = metrics.roc_curve(Y_new[test], predictrbl)  
            auc_1 = metrics.auc(fpr1,tpr1)
            auc_ = metrics.auc(fpr,tpr) 
            print("\nConfusiion Matrix Train: \n", metrics.confusion_matrix(Y_new[train], predictrbl1))
            print("\nConfusiion Matrix Test: \n", metrics.confusion_matrix(Y_new[test], predictrbl))
            akurasi.append(accuracy_score(Y_new[train], predictrbl1))
            akurasi.append(accuracy_score(Y_new[test], predictrbl))
            print("Akurasi Prediksi Train: ", accuracy_score(Y_new[train], predictrbl1))
            print("Akurasi Prediksi Test: ", accuracy_score(Y_new[test], predictrbl))
            print()
            print(metrics.classification_report(Y_new[train], predictrbl1))
            print(metrics.classification_report(Y_new[test], predictrbl))
            print("AUC Train: ", auc_1)
            print("AUC Test: ", auc_)
```

```python
#NBC 
kfold= StratifiedKFold(n_splits=4, shuffle=True)  
kfold.get_n_splits(X_new) 
kfold.get_n_splits(Y_new)
cl = BernoulliNB() 
akurasi = []

for train, test in kfold.split(X_new, Y_new): 
            (X_new[train], X_new[test])
            (Y_new[train], Y_new[test])
            hitung = cl.fit(X_new[train], Y_new[train])     
            predictnbc1 = cl.predict(X_new[train])
            predictnbc = cl.predict(X_new[test])
            fpr1, tpr1, _1 = metrics.roc_curve(Y_new[train], predictnbc1)
            fpr, tpr, _ = metrics.roc_curve(Y_new[test], predictnbc)  
            auc_1 = metrics.auc(fpr1,tpr1)
            auc_ = metrics.auc(fpr,tpr) 
            print("\nConfusiion Matrix Train: \n", metrics.confusion_matrix(Y_new[train], predictnbc1))
            print("\nConfusiion Matrix Test: \n", metrics.confusion_matrix(Y_new[test], predictnbc))
            akurasi.append(accuracy_score(Y_new[train], predictnbc1))
            akurasi.append(accuracy_score(Y_new[test], predictnbc))
            print("Akurasi Prediksi Train: ", accuracy_score(Y_new[train], predictnbc1))
            print("Akurasi Prediksi Test: ", accuracy_score(Y_new[test], predictnbc))
            print()
            print(metrics.classification_report(Y_new[train], predictnbc1))
            print(metrics.classification_report(Y_new[test], predictnbc))
            print("AUC Train: ", auc_1)
            print("AUC Test: ", auc_)
```

```python
#SVM Kernel Linear 
kfold= StratifiedKFold(n_splits=4, shuffle=True)  
kfold.get_n_splits(X_new) 
kfold.get_n_splits(Y_new)
lne= SVC(C=0.1, kernel='linear') 
akurasi = []

for train, test in kfold.split(X_new, Y_new): 
            (X_new[train], X_new[test])
            (Y_new[train], Y_new[test])
            hitung = lne.fit(X_new[train], Y_new[train])     
            predictlne1 = lne.predict(X_new[train])
            predictlne = lne.predict(X_new[test])
            fpr1, tpr1, _1 = metrics.roc_curve(Y_new[train], predictlne1)
            fpr, tpr, _ = metrics.roc_curve(Y_new[test], predictlne)  
            auc_1 = metrics.auc(fpr1,tpr1)
            auc_ = metrics.auc(fpr,tpr) 
            print("\nConfusiion Matrix Train: \n", metrics.confusion_matrix(Y_new[train], predictlne1))
            print("\nConfusiion Matrix Test: \n", metrics.confusion_matrix(Y_new[test], predictlne))
            akurasi.append(accuracy_score(Y_new[train], predictlne1))
            akurasi.append(accuracy_score(Y_new[test], predictlne))
            print("Akurasi Prediksi Train: ", accuracy_score(Y_new[train], predictlne1))
            print("Akurasi Prediksi Test: ", accuracy_score(Y_new[test], predictlne))
            print()
            print(metrics.classification_report(Y_new[train], predictlne1))
            print(metrics.classification_report(Y_new[test], predictlne))
            print("AUC Train: ", auc_1)
            print("AUC Test: ", auc_)
```

```python
#SVM Kernel RBF
kfold= StratifiedKFold(n_splits=4, shuffle=True)  
kfold.get_n_splits(X_new) 
kfold.get_n_splits(Y_new)
rbf= SVC(C=100, gamma=0.01, kernel='rbf')
akurasi = []
for train, test in kfold.split(X_new, Y_new): 
            (X_new[train], X_new[test])
            (Y_new[train], Y_new[test])
            hitung = rbf.fit(X_new[train], Y_new[train])     
            predictrbf1 = rbf.predict(X_new[train])
            predictrbf = rbf.predict(X_new[test])
            fpr1, tpr1, _1 = metrics.roc_curve(Y_new[train], predictrbf1)
            fpr, tpr, _ = metrics.roc_curve(Y_new[test], predictrbf)  
            auc_1 = metrics.auc(fpr1,tpr1)
            auc_ = metrics.auc(fpr,tpr) 
            print("\nConfusiion Matrix Train: \n", metrics.confusion_matrix(Y_new[train], predictrbf1))
            print("\nConfusiion Matrix Test: \n", metrics.confusion_matrix(Y_new[test], predictrbf))
            akurasi.append(accuracy_score(Y_new[train], predictrbf1))
            akurasi.append(accuracy_score(Y_new[test], predictrbf))
            print("Akurasi Prediksi Train: ", accuracy_score(Y_new[train], predictrbf1))
            print("Akurasi Prediksi Test: ", accuracy_score(Y_new[test], predictrbf))
            print()
            print(metrics.classification_report(Y_new[train], predictrbf1))
            print(metrics.classification_report(Y_new[test], predictrbf))
            print("AUC Train: ", auc_1)
            print("AUC Test: ", auc_)
```
